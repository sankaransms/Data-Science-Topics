## <font color="orange">**9.3) Groupby -> Filter - (like Group by and having in SQL)**</font>

- Grouped by Dataframe and on top of it perform aggregation and then filter on a column.

<font color="morange">**Q) Give the details of the movies by high budget directors:**</font>

- For finding the details of the movies by high budget directors lets assume, high budget director -> any director with **atleast one movie with budget >100M**  


### <font color="morange">**Lets solve the above question in three steps**</font>

**Learnings:**
- Use case of `names = data_dir_budget.loc[data_dir_budget["budget"] > 100,"director_name"]`
- Use case of `data.loc[data["director_name"].isin(names)]`


**Steps:**
- STEP 1: Group by -> aggregation
- STEP 2: From the grouped by dataframe -> Filtering
- STEP 3: To get the details of those movies directed by the "names"

```python
# STEP 1: Group by -> aggregation:
data_dir_budget = data.groupby("director_name")["budget"].max().reset_index()
data_dir_budget.head()
```

```python
# STEP 2: From the grouped by dataframe -> Filtering
names = data_dir_budget.loc[data_dir_budget["budget"] > 100,"director_name"]
names
```

```python
# STEP 3: To get the details of those movies directed by the "names"
data.loc[data["director_name"].isin(names)]
```

### <font color="morange">**Lets solve the above question in a single step using using groupby filter()**</font>

- Use case of `data.groupby("director_name").filter(lambda x: x["budget"].max() > 100)`

- Display only those rows where budget > 100 M after groupby("director_name") --> using groupby **filter()**


```python
# This filters all rows of those directors whose max budgt is greater than 100 million after grouped by a column:
data.groupby("director_name").filter(lambda x: x["budget"].max() > 100)
```

## <font color="orange">**9.4) Group based apply**</font>



<font color="morange">**Q) Categorize a movie as risky or non-risky, a movie is risky, if its budget is greater than the average revenue of all the movies of that director**</font>

```python
# Function to check risky investments
def check_risky(df):
    mean_revenue = df["revenue"].mean()
    risky_flag = df["budget"] > mean_revenue
    return risky_flag

# Apply the function and create the "risky" column
risky_flags = data.groupby("director_name").apply(lambda group: check_risky(group)).reset_index(level=0, drop=True)
data["risky"] = risky_flags
data
```
